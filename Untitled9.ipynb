```python
# =========================================================
# MONTHLY SARIMAX + LSTM HYBRID MODEL
# SEASONAL RAINFALL PREDICTION (PUBLISHABLE VERSION)
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# =========================================================
# 1. LOAD DATA
# =========================================================
df = pd.read_csv("chennai_crt.csv")
df["DATE"] = pd.to_datetime(df["DATE"], dayfirst=True)
df = df.sort_values("DATE")
df = df.set_index("DATE")

TARGET = "PRCP_mm"
df[TARGET] = pd.to_numeric(df[TARGET], errors="coerce")
df[TARGET] = df[TARGET].ffill().bfill()

# =========================================================
# 2. CONVERT DAILY → MONTHLY TOTAL RAINFALL
# =========================================================
df_month = df.resample("M").sum()

# OPTIONAL: Focus only Oct–Jan season
df_month = df_month[df_month.index.month.isin([10,11,12,1])]

# =========================================================
# 3. TRAIN / TEST SPLIT
# =========================================================
split_date = "2018-12-31"
train = df_month[df_month.index <= split_date].copy()
test  = df_month[df_month.index > split_date].copy()

# =========================================================
# 4. SARIMAX MODEL (SEASONAL = 4 MONTH CYCLE)
# =========================================================
sarimax = SARIMAX(
    train[TARGET],
    order=(1,1,1),
    seasonal_order=(1,1,1,4),   # 4-month seasonal cycle
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

train["SARIMAX_PRED"] = sarimax.predict(start=0, end=len(train)-1)
train["RESIDUAL"] = train[TARGET] - train["SARIMAX_PRED"]

test["SARIMAX_PRED"] = sarimax.predict(
    start=len(train),
    end=len(train)+len(test)-1
).set_axis(test.index)

# =========================================================
# 5. LSTM ON RESIDUALS
# =========================================================
residuals = train["RESIDUAL"].values.reshape(-1,1)

scaler = MinMaxScaler()
res_scaled = scaler.fit_transform(residuals)

def create_sequences(data, seq_length=6):
    X, y = [], []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

SEQ_LEN = 6  # 6 months memory
X_train, y_train = create_sequences(res_scaled, SEQ_LEN)

# =========================================================
# 6. LSTM MODEL
# =========================================================
model = Sequential([
    LSTM(32, return_sequences=True, input_shape=(SEQ_LEN,1)),
    LSTM(16),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")

early_stop = EarlyStopping(patience=10, restore_best_weights=True)

model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=8,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=0
)

# =========================================================
# 7. TEST RESIDUAL FORECAST
# =========================================================
combined_residuals = train["RESIDUAL"].tolist()
lstm_preds = []

for i in range(len(test)):
    seq_input = combined_residuals[-SEQ_LEN:]
    seq_scaled = scaler.transform(np.array(seq_input).reshape(-1,1))
    seq_scaled = seq_scaled.reshape(1,SEQ_LEN,1)

    pred_scaled = model.predict(seq_scaled, verbose=0)
    pred_residual = scaler.inverse_transform(pred_scaled)[0][0]

    lstm_preds.append(pred_residual)
    combined_residuals.append(pred_residual)

test["LSTM_RESIDUAL"] = lstm_preds

# =========================================================
# 8. FINAL HYBRID PREDICTION
# =========================================================
test["FINAL_PRED"] = test["SARIMAX_PRED"] + test["LSTM_RESIDUAL"]
test["FINAL_PRED"] = test["FINAL_PRED"].clip(lower=0)

# =========================================================
# 9. METRICS
# =========================================================
y_true = test[TARGET]
y_pred = test["FINAL_PRED"]

mae  = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2   = r2_score(y_true, y_pred)

print("\n===== MONTHLY SARIMAX + LSTM HYBRID PERFORMANCE =====")
print(f"MAE  : {mae:.3f}")
print(f"RMSE : {rmse:.3f}")
print(f"R²   : {r2:.3f}")

# =========================================================
# 10. VISUALIZATION
# =========================================================
plt.figure(figsize=(12,5))
plt.plot(test.index, y_true, marker="o", label="Actual")
plt.plot(test.index, y_pred, marker="s", label="Hybrid Prediction")
plt.title("Monthly Seasonal Rainfall Prediction (SARIMAX + LSTM)")
plt.xlabel("Date")
plt.ylabel("Monthly Rainfall (mm)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# =========================================================
# STACKED HYBRID MODEL (SARIMAX + LSTM + XGBOOST)
# OCT–JAN ONLY
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# =========================================================
# 1. LOAD DATA
# =========================================================
df = pd.read_csv("chennai_crt.csv")
df["DATE"] = pd.to_datetime(df["DATE"], dayfirst=True)
df = df.sort_values("DATE")
df = df.set_index("DATE")

TARGET = "PRCP_mm"
df[TARGET] = pd.to_numeric(df[TARGET], errors="coerce")
df[TARGET] = df[TARGET].ffill().bfill()

# =========================================================
# 2. FILTER OCT–JAN
# =========================================================
df = df[df.index.month.isin([10,11,12,1])]

# =========================================================
# 3. TRAIN TEST SPLIT
# =========================================================
split_date = "2018-12-31"
train = df[df.index <= split_date].copy()
test  = df[df.index > split_date].copy()

# =========================================================
# 4. SARIMAX MODEL
# =========================================================
sarimax = SARIMAX(
    train[TARGET],
    order=(1,1,1),
    seasonal_order=(1,1,1,4),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

train["SARIMAX"] = sarimax.predict(start=0, end=len(train)-1)
test["SARIMAX"]  = sarimax.predict(start=len(train), end=len(train)+len(test)-1).set_axis(test.index)

# =========================================================
# 5. XGBOOST MODEL
# =========================================================
for l in [1,2,3,7]:
    train[f"LAG_{l}"] = train[TARGET].shift(l)
    test[f"LAG_{l}"]  = df[TARGET].shift(l).loc[test.index]

features = ["LAG_1","LAG_2","LAG_3","LAG_7"]
train_ml = train.dropna()
test_ml  = test.dropna()

xgb = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=4,
    random_state=42
)

xgb.fit(train_ml[features], train_ml[TARGET])
test_ml["XGB"] = xgb.predict(test_ml[features])

# Align SARIMAX with XGB indices
test_ml["SARIMAX"] = test["SARIMAX"].loc[test_ml.index]

# =========================================================
# 6. LSTM MODEL (Direct Rainfall)
# =========================================================
scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(train[TARGET].values.reshape(-1,1))

def create_seq(data, seq_len=30):
    X,y=[],[]
    for i in range(len(data)-seq_len):
        X.append(data[i:i+seq_len])
        y.append(data[i+seq_len])
    return np.array(X), np.array(y)

SEQ_LEN=30
X_train_lstm, y_train_lstm = create_seq(scaled_train, SEQ_LEN)

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(SEQ_LEN,1)),
    LSTM(32),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.fit(X_train_lstm, y_train_lstm,
          epochs=40, batch_size=32, verbose=0,
          callbacks=[EarlyStopping(patience=5, restore_best_weights=True)])

# LSTM prediction for the test set
combined = train[TARGET].tolist()
lstm_preds=[]

for i in range(len(test)):
    seq = combined[-SEQ_LEN:]
    seq_scaled = scaler.transform(np.array(seq).reshape(-1,1)).reshape(1,SEQ_LEN,1)
    pred = model.predict(seq_scaled, verbose=0)
    pred_inv = scaler.inverse_transform(pred)[0][0]
    lstm_preds.append(pred_inv)
    combined.append(pred_inv)

test["LSTM"] = lstm_preds
test_ml["LSTM"] = test["LSTM"].loc[test_ml.index]

# =========================================================
# 7. STACKING (META MODEL)
# =========================================================
stack_features = ["SARIMAX","XGB","LSTM"]

# Generate LSTM predictions for the training data (for meta-model)
# These predictions correspond to train[TARGET].iloc[SEQ_LEN:]
train_lstm_predictions_scaled = model.predict(X_train_lstm, verbose=0)
train_lstm_predictions_raw = scaler.inverse_transform(train_lstm_predictions_scaled).flatten()

# Create a Series of LSTM predictions, aligning with the corresponding target index in 'train'
lstm_train_meta_feature_series = pd.Series(
    train_lstm_predictions_raw,
    index=train[TARGET].iloc[SEQ_LEN:].index
)

# Create a combined DataFrame for meta-model training
# This will handle leading NaNs in LSTM predictions (due to SEQ_LEN requirement)
# and ensures that features and target are perfectly aligned after dropping NaNs.
meta_train_df_for_fit = pd.DataFrame(index=train_ml.index)
meta_train_df_for_fit["SARIMAX"] = train["SARIMAX"].loc[train_ml.index]
meta_train_df_for_fit["XGB"] = pd.Series(xgb.predict(train_ml[features]), index=train_ml.index)
meta_train_df_for_fit["LSTM"] = lstm_train_meta_feature_series
meta_train_df_for_fit["TARGET"] = train_ml[TARGET]

# Drop any rows that have NaNs across any of the meta-features or the target
meta_train_df_for_fit = meta_train_df_for_fit.dropna()

meta = LinearRegression()
meta.fit(
    meta_train_df_for_fit[stack_features],
    meta_train_df_for_fit["TARGET"]
)

test_ml["FINAL_PRED"] = meta.predict(test_ml[stack_features])
test_ml["FINAL_PRED"] = test_ml["FINAL_PRED"].clip(lower=0)

# =========================================================
# 8. METRICS
# =========================================================
y_true = test_ml[TARGET]
y_pred = test_ml["FINAL_PRED"]

mae  = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
r2   = r2_score(y_true, y_pred)

print("\n===== STACKED HYBRID PERFORMANCE (OCT–JAN) =====")
print(f"MAE  : {mae:.3f}")
print(f"RMSE : {rmse:.3f}")
print(f"R²   : {r2:.3f}")

# =========================================================
# 9. PLOT
# =========================================================
plt.figure(figsize=(14,5))
plt.plot(test_ml.index, y_true, label="Actual")
plt.plot(test_ml.index, y_pred, label="Stacked Hybrid")
plt.legend()
plt.title("Stacked Hybrid Rainfall Prediction (Oct–Jan)")
plt.grid(True)
plt.show()
```
